```python
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
```

```python
# Mock API key - in real life you would get this from the service provider
API_KEY = 'dksfkjdsffdskfklsd'  # This is a fake key for demonstration
# Configure API requests
BASE_URL = "https://dog.ceo/api"
headers = {
    "Authorization": f"Bearer {API_KEY}",  # Example of how to include API key
    "Content-Type": "application/json"
}
```

```python
def make_authenticated_request(url):
    """Helper function to make requests with our API key"""
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None
```

```python
# 1. List all breeds (with mock authentication)
def list_all_breeds():
    print(" All Dog Breeds")
    data = make_authenticated_request(f"{BASE_URL}/breeds/list/all")
    if data:
        for breed, sub_breeds in data['message'].items():
            print(f" {breed.capitalize()}")
            if sub_breeds:
                print(f"    Sub-breeds: {', '.join(sub_breeds)}")
```

```python
# Run all examples
print(" Dog API Demo (with Mock Authentication)")
list_all_breeds()
```

```python
# 2. Get random image (with mock authentication)
def get_random_image():
    print("\n Random Dog Image")
    data = make_authenticated_request(f"{BASE_URL}/breeds/image/random")
    if data:
        display_image(data['message'])
```

```python
# Image display function
def display_image(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
get_random_image()
```

```python
# 3. Get images by breed
def get_breed_images(breed):
    print(f"\n Images of {breed.capitalize()}s")
    data = make_authenticated_request(f"{BASE_URL}/breed/{breed}/images")
    if data and data['message']:
        display_image(data['message'][0])
        print(f" Total images available: {len(data['message'])}")
```

```python
get_breed_images(breed='hound')
```
